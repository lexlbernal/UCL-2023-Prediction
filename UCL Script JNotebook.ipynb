```python
#Dividir dataframe
df_local= UCL_2007_2021[["Equipo Local", "GolesLocal","GolesVisitante"]]
df_visitante= UCL_2007_2021[["Equipo Visitante", "GolesLocal","GolesVisitante"]]


df_local= df_local.rename(columns={"Equipo Local": "Equipo","GolesLocal":"GolesAnotados","GolesVisitante":"GolesRecibidos"})
df_visitante= df_visitante.rename(columns={"Equipo Visitante": "Equipo","GolesLocal":"GolesRecibidos","GolesVisitante":"GolesAnotados"})

df_fortaleza_equipo=pd.concat([df_local,df_visitante], ignore_index=True).groupby("Equipo").mean()
pd.set_option('display.max_rows', None)
df_fortaleza_equipo
```

```python
#Funcion de prediccion de acuerdo a la distribución de poisson
#Condiciones para su uso:
#**Un evento dentro de este modelo esun gol anotado**
#1.Eventos se pueden contar
#2.Ocurrencia de eventos son independientes
#3.Tasa a la que ocurren los eventos es constante
#4.Dos eventos no pueden ocurrir exactamente en elmismo instante de tiempo


def predecir (Local,Visitante):
    if Local in df_fortaleza_equipo.index and Visitante in df_fortaleza_equipo.index:
        lamb_local=df_fortaleza_equipo.at[Local,"GolesAnotados"] * df_fortaleza_equipo.at[Visitante,"GolesRecibidos"]
        lamb_visitante=df_fortaleza_equipo.at[Visitante,"GolesAnotados"] * df_fortaleza_equipo.at[Local,"GolesRecibidos"]
        prob_local, prob_visitante,prob_empate=0,0,0
        for x in range(0,11): #goles equipo local
            for y in range (0,11):#goles equipo visitante
                p=poisson.pmf(x,lamb_local) * poisson.pmf(y,lamb_visitante)
                if x==y:
                    prob_empate +=p
                elif x > y:
                    prob_local +=p
                else:
                    prob_visitante +=p
        puntos_local=3*prob_local+prob_empate
        puntos_visitante=3*prob_visitante+prob_empate
        return(puntos_local,puntos_visitante)
    else:
        return(0,0)
    

```

```python
#Obteniendo grupos temporada 2022-2023
import pandas as pd

# Lee el archivo de Excel y convierte cada hoja en una tabla de Pandas
Fase_Grupos2022 = pd.read_excel('C:/Users/ENRIQUE LEON/Desktop/Data Science/Fase_Grupos2022_2023.xlsx', sheet_name=None)

# Imprime el nombre de cada hoja en el archivo de Excel
print(Fase_Grupos2022.keys())


# Accede a una hoja específica y conviértela en un DataFrame de Pandas
df_grupoA = Fase_Grupos2022['Grupo A']
df_grupoB = Fase_Grupos2022['Grupo B']
df_grupoC = Fase_Grupos2022['Grupo C']
df_grupoD = Fase_Grupos2022['Grupo D']
df_grupoE = Fase_Grupos2022['Grupo E']
df_grupoF = Fase_Grupos2022['Grupo F']
df_grupoG = Fase_Grupos2022['Grupo G']
df_grupoH = Fase_Grupos2022['Grupo H']


dict_tablas={}
dict_tablas["Grupo A"]=df_grupoA
dict_tablas["Grupo B"]=df_grupoB
dict_tablas["Grupo C"]=df_grupoC
dict_tablas["Grupo D"]=df_grupoD
dict_tablas["Grupo E"]=df_grupoE
dict_tablas["Grupo F"]=df_grupoF
dict_tablas["Grupo G"]=df_grupoG
dict_tablas["Grupo H"]=df_grupoH
print(dict_tablas.keys())
print(dict_tablas.values())
```

```python
###Lamando al fixture 2022
df_fixture=pd.read_excel("C:/Users/ENRIQUE LEON/Desktop/Data Science/Arreglo_UCL_2022.xlsx")
df_fixture

df_fixture_fasegrupos_96=df_fixture[:96].copy()
df_fixture_octavos=df_fixture[96:104].copy()
df_fixture_cuartos=df_fixture[104:108].copy()
df_fixture_semis=df_fixture[108:110].copy()
df_fixture_final=df_fixture[110:111].copy()

for grupo in dict_tablas:
    equipos_por_grupo = dict_tablas[grupo] ["Equipo"].values
    df_fixture_group_6=df_fixture_fasegrupos_96[df_fixture_fasegrupos_96["Local"].isin(equipos_por_grupo)]
    for index, row in df_fixture_group_6.iterrows():
        Local, Visita=row["Local"], row["Visita"]
        puntos_local, puntos_visita=predecir(Local, Visita)
        dict_tablas[grupo].loc[dict_tablas[grupo]["Equipo"]==Local, "PUNTOS"]+=puntos_local
        dict_tablas[grupo].loc[dict_tablas[grupo]["Equipo"]==Visita, "PUNTOS"]+=puntos_visita

    dict_tablas[grupo]=dict_tablas[grupo].sort_values("PUNTOS", ascending=False).reset_index()
    dict_tablas[grupo]=dict_tablas[grupo][["Equipo","PUNTOS"]]
    dict_tablas[grupo]=dict_tablas[grupo].round(0)
```

```python
#Definiendo octavos

for grupo in dict_tablas:
    primero_grupo=dict_tablas[grupo].loc[0,"Equipo"]
    segundo_grupo=dict_tablas[grupo].loc[1,"Equipo"]
    df_fixture_octavos.replace({f"Segundo {grupo}":segundo_grupo,
                                f"Primero {grupo}":primero_grupo},inplace=True)
    
df_fixture_octavos["Ganador"]="?"
df_fixture_octavos
```

```python
#Definiendo ganador de octavos
def definir_ganador(df_fixture_actualizado):
    for index, row in df_fixture_actualizado.iterrows():
        Local, Visita=row["Local"], row["Visita"]
        puntos_local, puntos_visita=predecir(Local, Visita)
        if puntos_local > puntos_visita:
            Ganador=Local
        else:
            Ganador=Visita
        df_fixture_actualizado.loc[index, "Ganador"] = Ganador
    return df_fixture_actualizado

definir_ganador(df_fixture_octavos)
```

```python
#Definiendo Cuartos de final

def actualizar_tabla(df_fixture_round_1,df_fixture_round_2):
    for index, row in df_fixture_round_1.iterrows():
        Ganador=df_fixture_round_1.loc[index,"Ganador"]
        Partido=df_fixture_round_1.loc[index,"Resultado"]
        df_fixture_round_2.replace({f"Ganador {Partido}":Ganador}, inplace=True)
    df_fixture_round_2["Ganador"]="?"
    return df_fixture_round_2


actualizar_tabla(df_fixture_octavos,df_fixture_cuartos)
definir_ganador(df_fixture_cuartos)
```
